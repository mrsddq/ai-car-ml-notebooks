```python
import matplotlib.pyplot as plt

import matplotlib.dates as mdates

from datetime import datetime

 

# Event dates and corresponding unlisted share prices

dates = [

    datetime(2024, 9, 1),   # Before rally (low base)

    datetime(2024, 12, 24), # Funding by Rainmatter etc.

    datetime(2025, 3, 27),  # SEBI expiry rule change

    datetime(2025, 7, 8),   # ₹1,000 Cr raise approved

    datetime(2025, 8, 7),   # Reappointment of CEO

]

 

prices = [2, 13.5, 8, 5, 5]  # Approximate unlisted prices at those times

 

# Labels for annotation

labels = [

    "Before rally (~₹1-2)",

    "₹238 Cr funding (~₹13-14)",

    "SEBI expiry rule (~₹8)",

    "₹1,000 Cr raise approved",

    "CEO reappointed (~₹5)"

]

 

# Create the plot

plt.figure(figsize=(12, 6))

plt.plot(dates, prices, marker='o', linestyle='-', color='dodgerblue', linewidth=2)

 

# Annotate each point

for i, label in enumerate(labels):
    offset = 20 if prices[i] < max(prices) else -30
    plt.annotate(
            label,
            (dates[i], prices[i]),
            textcoords="offset points",
            xytext=(0, offset),
            ha='center',
            fontsize=9,
            rotation=90,
            va='bottom' if offset > 0 else 'top'
        )
 

# Formatting

plt.title("MSEI Unlisted Share Price vs Key Events (2024–2025)", fontsize=12)

plt.xlabel("Date", fontsize=10)

plt.ylabel("Unlisted Share Price (INR)", fontsize=10)

plt.grid(True)

plt.xticks(rotation=45)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

 

# Save the image

plt.tight_layout()

plt.savefig("msei_unlisted_share_price_chart.png")

plt.show()
```
